# 비율 검정으로 임계값 설정

- two sample z-test
    - 연속형 변수에 대해 모든 값을 10씩 증가시켜가며 값을 기준으로 target attribute인 'BMI'의 비율이 가장 크게 차이가 나는 지점을 찾는다.
    - p-value는 0.05보다 작은 값을 할당
    
#### 분석 플로우

1. p-value 0.0001 이하
    - 두 그룹 간의 차이 1000 이하
    - 없다면 2000 이하
    - 다시 없다면 3000 이하에서
    - 다시 없다면 4000 이하에서
    - 다시 없다면 5000 이하에서
    - 다시 없다면 6000 이하에서
    
2. p-value 0.001 이하
    - 차이 1000 이하
    - 없다면 2000 이하
    - 다시 없다면 3000 이하에서
    - 다시 없다면 4000 이하에서
    - 다시 없다면 5000 이하에서
    - 다시 없다면 6000 이하에서
    
3. p-value 0.01 이하
    - 차이 1000 이하
    - 없다면 2000 이하
    - 다시 없다면 3000 이하에서
    - 다시 없다면 4000 이하에서
    - 다시 없다면 5000 이하에서
    - 다시 없다면 6000 이하에서
    - 다시 없다면 7000 이하
    - 다시 없다면 8000 이하


```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest
columns=['PA_SWD_S','PA_SWD_N','PA_SWK_S','PA_SWK_N','SleepAvg_WD','SleepAvg_WK'] #연속형 변수들

for col in columns:
    ratio_big = []
    ratio_same = []
    try:
        print("=============p-value 0.0001 이하===============")
        for i in range(1,int(max(df[col])),10):
            group1 = df[df[col] < i]
            group2 = df[df[col] >= i]
            diff = abs(len(group1)-len(group2))
            if diff > 10000:
                pass
            else:
                bmi_1 = group1['BMI'].value_counts()
                bmi_2 = group2['BMI'].value_counts()
                if len(bmi_1) == 1:
                    ratio1 = bmi_1.index[0]
                else:
                    ratio1 = bmi_1[1]/bmi_1[0]
                if len(bmi_2) == 1:
                    ratio2 = bmi_2.index[0]
                else:
                    ratio2 = bmi_2[1]/bmi_2[0]
                count = np.array([bmi_1[1], bmi_2[1]])
                nobs = np.array([len(group1), len(group2)])
                z, p_value = proportions_ztest(count=count, nobs=nobs, value=0)
        #         print("임계값:",i, "기준")
        #         print("group1 사이즈:",len(group1), "  group2 사이즈:",len(group1), "group1 비율" , ratio1, "  group2 비율" ,ratio2, p_value)
                alpha = 0.0001
                if p_value > alpha :
                    ratio_same.append((len(group1), len(group2), abs(len(group1)-len(group2)), i, p_value))
                else :
                    ratio_big.append((len(group1), len(group2), abs(len(group1)-len(group2)), i,p_value))
        p_value_df = pd.DataFrame(ratio_big, columns=['group1 size','group2 size','2개 그룹 크기 차이','임계값','p-value'])

        try:
            print('두 그룹 간 1000개 이하 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<1000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0], '==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<1000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
        except:
            try:
                print('두 그룹 간 2000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<2000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0], '==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<2000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
            except:
                try:         
                    print('두 그룹 간 3000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<3000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<3000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
                except:
                    try:
                        print('두 그룹 간 4000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<4000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<4000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
                    except:
                        try:
                            print('두 그룹 간 5000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<5000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<5000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
                        except:
                            print('두 그룹 간 6000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<6000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<6000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
    except:
        try:
            ratio_big = []
            ratio_same = []
            print("=============p-value 0.001 이하===============")
            for i in range(1,int(max(df[col])),10):
                group1 = df[df[col] < i]
                group2 = df[df[col] >= i]
                diff = abs(len(group1)-len(group2))
                if diff > 10000:
                    pass
                else:
                    bmi_1 = group1['BMI'].value_counts()
                    bmi_2 = group2['BMI'].value_counts()
                    if len(bmi_1) == 1:
                        ratio1 = bmi_1.index[0]
                    else:
                        ratio1 = bmi_1[1]/bmi_1[0]
                    if len(bmi_2) == 1:
                        ratio2 = bmi_2.index[0]
                    else:
                        ratio2 = bmi_2[1]/bmi_2[0]
                    count = np.array([bmi_1[1], bmi_2[1]])
                    nobs = np.array([len(group1), len(group2)])
                    z, p_value = proportions_ztest(count=count, nobs=nobs, value=0)
            #         print("임계값:",i, "기준")
            #         print("group1 사이즈:",len(group1), "  group2 사이즈:",len(group1), "group1 비율" , ratio1, "  group2 비율" ,ratio2, p_value)
                    alpha = 0.001
                    if p_value > alpha :
                        ratio_same.append((len(group1), len(group2), abs(len(group1)-len(group2)), i, p_value))
                    else :
                        ratio_big.append((len(group1), len(group2), abs(len(group1)-len(group2)), i,p_value))
            p_value_df = pd.DataFrame(ratio_big, columns=['group1 size','group2 size','2개 그룹 크기 차이','임계값','p-value'])

            try:
                print('두 그룹 간 1000개 이하 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<1000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<1000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
            except:
                try:
                    print('두 그룹 간 2000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<2000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<2000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
                except:
                    try:         
                        print('두 그룹 간 3000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<3000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<3000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
                    except:
                        try:
                            print('두 그룹 간 4000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<4000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<4000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
                        except:
                            try:
                                print('두 그룹 간 5000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<5000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<5000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
                            except:
                                print('두 그룹 간 6000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<6000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<6000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])

        except:
            print("=============p-value 0.01 이하===============")
            ratio_big = []
            ratio_same = []
            for i in range(1,int(max(df[col])),10):
                group1 = df[df[col] < i]
                group2 = df[df[col] >= i]
                diff = abs(len(group1)-len(group2))
                if diff > 10000:
                    pass
                else:
                    bmi_1 = group1['BMI'].value_counts()
                    bmi_2 = group2['BMI'].value_counts()
                    if len(bmi_1) == 1:
                        ratio1 = bmi_1.index[0]
                    else:
                        ratio1 = bmi_1[1]/bmi_1[0]
                    if len(bmi_2) == 1:
                        ratio2 = bmi_2.index[0]
                    else:
                        ratio2 = bmi_2[1]/bmi_2[0]
                    count = np.array([bmi_1[1], bmi_2[1]])
                    nobs = np.array([len(group1), len(group2)])
                    z, p_value = proportions_ztest(count=count, nobs=nobs, value=0)
            #         print("임계값:",i, "기준")
            #         print("group1 사이즈:",len(group1), "  group2 사이즈:",len(group1), "group1 비율" , ratio1, "  group2 비율" ,ratio2, p_value)
                    alpha = 0.01
                    if p_value > alpha :
                        ratio_same.append((len(group1), len(group2), abs(len(group1)-len(group2)), i, p_value))
                    else :
                        ratio_big.append((len(group1), len(group2), abs(len(group1)-len(group2)), i,p_value))
            p_value_df = pd.DataFrame(ratio_big, columns=['group1 size','group2 size','2개 그룹 크기 차이','임계값','p-value'])

            try:
                print('두 그룹 간 1000개 이하 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<1000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<1000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
            except:
                try:
                    print('두 그룹 간 2000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<2000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<2000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
                except:
                    try:         
                        print('두 그룹 간 3000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<3000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<3000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
                    except:
                        try:
                            print('두 그룹 간 4000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<4000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<4000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
                        except:
                            try:
                                print('두 그룹 간 5000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<5000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<5000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
                            except:
                                try:
                                    print('두 그룹 간 6000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<6000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<6000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
                                except:
                                    try:
                                        print('두 그룹 간 7000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<7000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<7000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
                                    except:
                                        print('두 그룹 간 8000개 차이가 나는 임계값 수 :', len(p_value_df),'      ', col,'에 대한 임계값:',p_value_df[p_value_df['2개 그룹 크기 차이']<8000].sort_values('p-value',ascending=True).reset_index(drop=True)['임계값'][0],'==p-value:', p_value_df[p_value_df['2개 그룹 크기 차이']<8000].sort_values('p-value',ascending=True).reset_index(drop=True)['p-value'][0])
                                
                                        
        
            
```

# 다중공선성 체크
## (다중공선성 높은 변수 1개씩 제거하면서 VIF가 10미만일 때 stop)

- 즉, VIF가 높은 순서대로 하나씩 제거하면서 다시 VIF 도출 (10 미만일때까지)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 다중공선성 확인
def check_vif(df):
    global vif
    vif = pd.DataFrame()
    vif["VIF Factor"] = [variance_inflation_factor(
        df.values, i) for i in range(df.shape[1])]
    vif["features"] = df.columns
```

```python
check_vif(df)
```

```python
for i in range(1000):
    check_vif(df)
    if len(vif[vif['VIF Factor'] >=10]) >=1:
        remove_col = vif[vif['VIF Factor'] >=10].sort_values('VIF Factor', ascending=False).iloc[0,1]
        df.drop(remove_col, axis=1, inplace=True)
        print('제거된 변수는', remove_col)
    else:
        display(vif)
        break
```

```python
df.reset_index(drop=True, inplace=True)
```
