# 결측치 처리

1. 설문조사 특성상 모름, 무응답 (9999,8888,999,888 등) 처리
2. 결측치 10% 이상인 열 제거
3. 나머지 결측행 삭제

```python
import numpy as np
for i in list(df.columns):
    df[i] = df[i].replace({9999:np.nan})
    df[i] = df[i].replace({8888:np.nan})
    df[i] = df[i].replace({999:np.nan})
    df[i] = df[i].replace({888:np.nan})
    df[i] = df[i].replace({99:np.nan})
    df[i] = df[i].replace({88:np.nan})
```

```python
# df에 있는 변수들 중 변수값이 9인 거를 포함하는 변수 확인
count_9 = 0
list_9 = []
for i in list(df.columns):
    if 9 in df[i].value_counts():
        count_9 += 1
        list_9.append(i)

print(count_9)
print(list_9)

# df에 있는 변수들 중 변수값이 8인 거를 포함하는 변수 확인
count_8 = 0
list_8 = []
for i in list(df.columns):
    if 8 in df[i].value_counts():
        count_8 += 1
        list_9.append(i)

print(count_8)
print(list_9)
```

# 다중공선성 체크
## (다중공선성 높은 변수 1개씩 제거하면서 VIF가 10미만일 때 stop)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 다중공선성 확인
def check_vif(df):
    global vif
    vif = pd.DataFrame()
    vif["VIF Factor"] = [variance_inflation_factor(
        df.values, i) for i in range(df.shape[1])]
    vif["features"] = df.columns
```

```python
for i in range(1000):
    check_vif(df)
    if len(vif[vif['VIF Factor'] >=10]) >=1:
        remove_col = vif[vif['VIF Factor'] >=10].sort_values('VIF Factor', ascending=False).iloc[0,1]
        df.drop(remove_col, axis=1, inplace=True)
        print('제거된 변수는', remove_col)
    else:
        display(vif)
        break
```

```python
df.reset_index(drop=True, inplace=True)
```
